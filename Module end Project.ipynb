# How many are there in each Team and the percentage splitting with respect to the total employees.

```python
#calcuculate the number of teams
number_of_teams=df['Team'].value_counts()

percentage_split=(number_of_teams/458)*100

new_df = pd.DataFrame({
    'Team': number_of_teams.index,
    'Number of Players': number_of_teams.values,
    'Percentage Split': percentage_split.values
})
new_df
```

# 3.Find from which age group most of the employees belong to.


```python
age_grp=pd.cut(df['Age'],bins=[18,25,35,45,55,65])
age_grp_counts=age_grp.value_counts()
print(age_grp_counts) 
```

# 4.Find out under which team and position, spending in terms of salary is high.

```python
data=df.groupby(['Team','Position'])['Salary'].sum()
max_scalary=data.idxmax()
A,B=max_scalary
print("Team which belongs to spending maximum salary is;",A)
print("position which belongs to spending maximum salary is;",B)
```

# 5.Find if there is any correlation between age and salary , represent it visually.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(df, x="Age", y="Salary")
plt.title("Relationship between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
